```python
import pandas as pd
import numpy as np
from keras.layers.core import Dense, Activation, Dropout
from keras.layers.recurrent import LSTM
from keras.models import Sequential
import time
from sklearn.metrics import mean_squared_error
from matplotlib import pyplot
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
series = pd.read_csv('sine-wave.csv', header = None)
```

```python
power_series = pd.read_csv('model_data.csv')
power_series = pd.DataFrame(power_series['DALMP'])
power_series.head()
```

```python
power_series.shape
```

```python
pyplot.figure(figsize =(20,6))
pyplot.plot(power_series.values)
pyplot.show()
```

```python
pyplot.figure(figsize=(20,6))
pyplot.plot(power_series.values[:50])
pyplot.show()
```

```python
scaler = MinMaxScaler(feature_range =(-1,1))
scaled = scaler.fit_transform(power_series.values)
series = pd.DataFrame(scaled)
series.head()
```

```python
window_size = 150

series_s = series.copy()
for i in range(window_size):
    series = pd.concat([series, series_s.shift(-(i+1))], axis = 1)
    
series.dropna(axis = 0, inplace = True)
```

```python
series.head()
```

```python
series.shape
```

```python
nrow = round(0.8*series.shape[0])
```

```python
train = series.iloc[:nrow, :]
test = series.iloc[nrow:, :]
```

```python
from sklearn.utils import shuffle
train = shuffle(train)
```

```python
n_out = 24

train_X = train.iloc[:, :-n_out]
train_y = train.iloc[:, -n_out:]
test_X = test.iloc[:, :-n_out]
test_y = test.iloc[:, -n_out:]
```

```python
train_X = train_X.values
train_y = train_y.values
test_X = test_X.values
test_y = test_y.values
```

```python
print(train_X.shape)
print(train_y.shape)
print(test_X.shape)
print(test_y.shape)
```

```python
train_y[0:5]
```

```python
train_X = train_X.reshape(train_X.shape[0],train_X.shape[1],1)
test_X = test_X.reshape(test_X.shape[0],test_X.shape[1],1)
```

```python
train_X.shape
```

```python
input_dim = window_size - n_out + 1
input_dim
```

```python
# Define the LSTM model
model = Sequential()
model.add(LSTM(input_shape = (input_dim, 1), output_dim = input_dim, return_sequences = True))
model.add(Dropout(0.5))
model.add(LSTM(256))
model.add(Dropout(0.5))
model.add(Dense(n_out))
model.add(Activation("linear"))
model.compile(loss="mse", optimizer="adam")
model.summary()
```

```python
start = time.time()
model.fit(train_X,train_y,batch_size=512,nb_epoch=3,validation_split=0.1)
print("> Compilation Time : ", time.time() - start)
```

```python
preds = model.predict(test_X)
preds = scaler.inverse_transform(preds)
actuals = scaler.inverse_transform(test_y.reshape(-1, 1))
```

```python
pyplot.plot(actuals[:24])
pyplot.plot(preds[0:24])
pyplot.show()
```

```python
mean_squared_error(actuals,preds)
```
