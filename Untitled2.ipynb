```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
from keras.models import Sequential
from keras.layers import Dense
from keras.layers.recurrent import LSTM
```

```python
data_path = "C:/Users/cmpgervais/Documents/power-price-forecasting/model_data.csv"

model_data = pd.read_csv(data_path)
```

```python
model_data.head()
```

```python
model_data = model_data.drop(['month'], axis = 1)
```

```python
dummies = pd.get_dummies(model_data['hour'], prefix = 'hour', drop_first = False)
model_data = pd.concat([model_data, dummies], axis = 1)
model_data = model_data.drop(['hour'], axis = 1)
```

```python
model_data.head()
```

```python
quant_features = ['RTLMP', 'LOAD', 'WIND', 'OUTAGES', 'lag24', 'lag48', 'lag72', 'lag96', 'lag120', 'lag144', 'lag168']
# Store scalings in a dictionary so we can convert back later
scaled_features = {}
for each in quant_features:
    mean, std = model_data[each].mean(), model_data[each].std()
    scaled_features[each] = [mean, std]
    model_data.loc[:, each] = (model_data[each] - mean)/std
```

```python
model_data.head()
```

```python
# Save data for approximately the last 21 days 
test_data = model_data[-90*24:]

# Now remove the test data from the data set 
model_data = model_data[:-90*24]

# Separate the data into features and targets
target_fields = ['RTLMP']
features, targets = model_data.drop(target_fields, axis=1), model_data[target_fields]
test_features, test_targets = test_data.drop(target_fields, axis=1), test_data[target_fields]

train_features, train_targets = features[:-150*24], targets[:-150*24]
val_features, val_targets = features[-150*24:], targets[-150*24:]
```

```python
n_cols = train_features.shape[1]
```

```python
model = Sequential()
model.add(Dense(600, activation = 'relu', input_shape = (n_cols,)))
model.add(Dense(300, activation = 'relu'))
model.add(Dense(100, activation = 'relu'))
model.add(Dense(1))
model.compile(loss = 'mean_squared_error', optimizer = 'adam', metrics = ['mean_squared_error'])
```
